```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
#from wordcloud import WordCloud
```

```python
nltk.download('vader_lexicon')
nltk.download('stopwords')
```

```python
sia=SentimentIntensityAnalyzer()
```

```python
def preprocess_text(text):
  tokens=word_tokenize(text.lower())
  stop_words=set(stopwords.words('english'))
  tokens=[token for token in tokens if token not in stop_words and token not in string.punctuation and not token.isdigit()]
  return ' '.join(tokens)
```

```python
nltk.download('punkt')
```

```python
data['reviews.text']=data['reviews.text'].fillna('')
```

```python
data['reviews.text']=data['reviews.text'].apply(preprocess_text)
```

```python
def is_fake(review):
  sentiment_score=sia.polarity_scores(review)['compound']
  if sentiment_score < -0.5 or sentiment_score > 0.5:
    return False
  return True
```

```python
data['reviews.text']=data['reviews.text'].fillna('')
data['is_fake']=data['reviews.text'].apply(is_fake)
```

```python
fake_reviews=data[data['is_fake']==True].shape[0]
genuine_reviews=data[data['is_fake']==False].shape[0]
```

```python
print(f"Number of fake reviews: {fake_reviews}")
print(f"Number of genuine reviews: {genuine_reviews}")
```

```python
data['label']=data['is_fake'].apply(lambda x: 'Fake' if x else 'Genuine')
for label, group in data.groupby('label'):
  print(f"{label} reviews:\n")
  for review in group['reviews.text'].head():
    print(review)
    print('--------------')
  print()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
X=data['reviews.text']
Y=data['is_fake']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
vectorizer=TfidfVectorizer()
```

```python
X_train_vec=vectorizer.fit_transform(X_train)
```

```python
X_test_vec=vectorizer.transform(X_test)
```

```python
rf_classifier=RandomForestClassifier(n_estimators=100,random_state=42)
rf_classifier.fit(X_train_vec,Y_train)
```

```python
dt_classifier=DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train_vec,Y_train)
```

```python
rf_pred=rf_classifier.predict(X_test_vec)
rf_accuracy=accuracy_score(Y_test,rf_pred)
print(f"Random Forest Accuracy: {rf_accuracy}")
```

```python
dt_pred=dt_classifier.predict(X_test_vec)
dt_accuracy=accuracy_score(Y_test,dt_pred)
print(f"Decision Tree Accuracy: {dt_accuracy}")
```
